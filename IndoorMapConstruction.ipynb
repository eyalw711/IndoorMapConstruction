### Step 1: Build a building

### Step 2: Let's walk around inside
You have a phone in your pocket, with revolutionary indoor location technology.  
Select fix periodicity:

```python
w = IntSlider(value=2,
    min=1,
    max=3,
    step=1,
    description='Seconds per fix:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='i',
    slider_color='white')
display(w)
```

A good choice.

```python
from TrajectoryMaker import TrajectoryMaker
secondsPerFix = w.value

# This is a trajectory maker
trajectoryMaker = TrajectoryMaker(bd)

# This is our trajectory
trajectory = trajectoryMaker.makeFixList(secondsPerFix)

# Plot our trajectory inside the building
plt.plot(xs, ys)
txs, tys = trajectory.scatterXY()
plt.plot(txs, tys)
```
